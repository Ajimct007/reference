```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the train data set
df_train = pd.read_csv('D:\\Data Science\\Krish_Naik\\EDA\\02_Live_EDA\\Black_Friday\\train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('D:\\Data Science\\Krish_Naik\\EDA\\02_Live_EDA\\Black_Friday\\test.csv')
df_test.head()
```

### Merging the both train and test dataset

```python
df=df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'], axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
#Handling the categorical feature Gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
#Handling the categorical feature Age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1, '55+':7, '26-35':3, '46-50':5, '51-55':6, '36-45':4, '18-25':2})
df['Age'].head()
```

```python
##fixing categorical City_categort
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
##drop City Category Feature since we fixed
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#focus on replacing missing values
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
## Replace the missing value with mode
df['Product_Category_2'].mode()[0]
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
## Replace the missing values with mode
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```
