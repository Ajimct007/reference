```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Assuming you have a GeoDataFrame named 'districts' with a 'Dist_Name' column
# and you want to highlight the top 5 districts based on some criterion (e.g., population)

# Step 3: Create a list of colors for the top 5 districts
top_5_colors = ['red', 'green', 'blue', 'purple', 'orange', 'red', 'green', 'blue', 'purple', 'orange']

# Step 4: Create a plot with the same color for all districts
fig, ax = plt.subplots(figsize=(10, 8))
districts.plot(ax=ax, color='lightgray', edgecolor='black', label='Other Districts')

# Step 5: Plot the top 5 districts with different colors and labels
for idx, row in top_5_districts.iterrows():
    top_5_districts[top_5_districts.index == idx].plot(ax=ax, color=top_5_colors[idx], edgecolor='black', label=f"{idx+1}: {row['Dist_Name']}")

    # Annotate the top 5 districts with their numbers and names
    plt.annotate(text=f"{idx+1}: {row['Dist_Name']}",
                 xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                 xytext=(3, 3),
                 textcoords="offset points",
                 fontsize=8,
                 color='black')

# Step 6: Remove axis labels
ax.set_xticks([])
ax.set_yticks([])

# Step 7: Create custom legend entries
legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor=top_5_colors[i], markersize=10, label=f"{i+1}: {name}")
                   for i, name in enumerate(top_5_districts['Dist_Name'])]

# Add a legend outside the map
ax.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 1))

# You can add additional elements to the plot, such as title, labels, etc., as needed.
plt.title("Districts with Top 5 Population")
plt.show()
```

```python
# Step 3: Merge the GeoDataFrame with the top 5 districts and custom colors
districts_with_colors = districts.merge(top_5_districts[['Dist_Name']], on='Dist_Name', how='left')
districts_with_colors['highlight'] = districts_with_colors['Dist_Name'].isin(top_5_districts['Dist_Name'])

# Step 4: Create a choropleth map with highlighted districts
fig = px.choropleth(districts_with_colors,
                    geojson=districts_with_colors['geometry'],
                    locations=districts_with_colors.index,
                    color='highlight',
                    color_discrete_map={True: 'red', False: 'lightgray'},
                    projection="mercator")

# Step 5: Update layout properties
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(title="Districts with Top 5 Population", margin={"r": 0, "t": 50, "l": 0, "b": 0}, width=800, height=600)

# Show the map
fig.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Replace 'path_to_telangana_shapefile.shp' with the actual path to the Telangana shapefile
telangana_shapefile_path = 'D:\\Data Science\\Code basics\\Challenge\\Challenge_7\\Telangana_Shape_Files\\TS_DISTRICT_BOUNDARIES_33.shp'

# Read the Telangana shapefile using GeoPandas
telangana_gdf = gpd.read_file(telangana_shapefile_path)

# Plot the Telangana boundary with a margin
margin = 0.1  # Adjust the margin as needed
x_min, y_min, x_max, y_max = telangana_gdf.total_bounds
x_margin = (x_max - x_min) * margin
y_margin = (y_max - y_min) * margin

fig, ax = plt.subplots(figsize=(8, 8))
telangana_gdf.boundary.plot(ax=ax, color='blue', linewidth=2)

# Set the plot limits with a margin
ax.set_xlim(x_min - x_margin, x_max + x_margin)
ax.set_ylim(y_min - y_margin, y_max + y_margin)

# Annotate each district boundary with its name
for _, row in telangana_gdf.iterrows():
    district_name = row['Dist_Name']  # Replace with the actual column name
    district_center = row['geometry'].centroid
    ax.annotate(district_name, (district_center.x, district_center.y),
                fontsize=10, color='black', ha='center')

# Remove the axes
ax.set_xticks([])
ax.set_yticks([])

# Show the plot
plt.show()
```
